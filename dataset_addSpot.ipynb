### subway

```python
subway=pd.read_csv("data/subway_addSpot.csv", encoding="euc-kr")
```

```python
subway.head(2)
```

```python
subway.columns=["unnamed","spot_num","date","subway_input","subway_output","subway_sum"]
```

```python
# 정규표현식
def date_change(x):  
    pattern=re.compile("-")
    date=re.sub(pattern,"",x)
    return(date)
```

```python
subway["date"]=subway["date"].map(date_change)
```

```python
subway.head(2)
```

```python
subway["date"]=subway["date"].astype(float)
```

```python
subway.info()
```

```python
subway_m=subway.drop("unnamed",axis=1)
```
